# Exercise 07: Hidden word in sound

It is possible to hide message in sound by manipulating the spectrogram. Some examples can be found in <a href="https://twistedsifter.com/2013/01/hidden-images-embedded-into-songs-spectrographs/">here</a>. The following is an example to draw a trapezium in spectrogram.

```python
%matplotlib inline

import sys
sys.path.append('../ThinkDsp')
import thinkdsp
import thinkplot
```

```python
signal1 = thinkdsp.Chirp(start=200, end=1400)
wave1 = signal1.make_wave(duration=2)
```

```python
signal2 = thinkdsp.Chirp(start=1400, end=200)
wave2 = signal2.make_wave(duration=2)
```

```python
wave3 = thinkdsp.sin_wave(freq=200, duration=6)
```

```python
wave4 = thinkdsp.sin_wave(freq=1400, duration=2)
```

```python
wave5 = thinkdsp.rest(2)
```

```python
wave = wave1 | wave4 | wave2
wave = wave + wave3
wave = wave5 | wave | wave5
wave.make_audio()
```

```python
wave.plot()
```

```python
sg = wave.make_spectrogram(seg_length=4096)
sg.plot(high=2500, shading='auto')
```
