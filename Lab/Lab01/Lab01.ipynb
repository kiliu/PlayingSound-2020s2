# Lab 01: Sound generation and processing

```python
%matplotlib inline

import sys
sys.path.append('../ThinkDsp')
import thinkdsp
import thinkplot
```

## Generate simple sinusoid wave 

```python
wave_440 = thinkdsp.sin_wave(freq=440, duration=1) 
wave_440.make_audio()
```

```python
wave_440.plot()
```

```python
wave_440_segment = wave_440.segment(start=0, duration=0.01)
wave_440_segment.plot()
```

```python
wave_880 = thinkdsp.sin_wave(freq=880, duration=1)  
wave_880.make_audio()
```

```python
wave_880_segment = wave_880.segment(start=0, duration=0.01)
wave_880_segment.plot()
```

## Combine sound waves

```python
comb_wave = wave_440 | wave_880
comb_wave.make_audio()
```

## Combine sound waves with better quality

```python
wave_440.apodize()
wave_880.apodize()
comb_wave = wave_440 | wave_880
comb_wave.make_audio()
```

```python
wave_440_segment = wave_440.segment(start=0, duration=0.01)
wave_440_segment.plot()
```

## Mix sound waves

```python
wave_A3 = thinkdsp.sin_wave(freq=220.00, duration=2) 
wave_E4 = thinkdsp.sin_wave(freq=329.63, duration=2) 

mix_wave = wave_A3 + wave_E4
mix_wave.make_audio()
```

```python
mix_segment = mix_wave.segment(start=0, duration=0.03)
mix_segment.plot()
```

## Mix sound waves with better quality

```python
mix_wave = wave_A3 + wave_E4
mix_wave.normalize()
mix_wave.make_audio()
```

```python
mix_segment = mix_wave.segment(start=0, duration=0.03)
mix_segment.plot()
```

## Spectrum of sound wave

```python
sp = mix_wave.make_spectrum()
sp.plot(high=500)
```

## Generate sound wave from the the square signal

```python
square_sig = thinkdsp.SquareSignal(freq=200, amp=1, offset=0)
square_wave = square_sig.make_wave(duration=1.5, start=0, framerate=11025)
square_wave.make_audio()
```

```python
sq_segment = square_wave.segment(start=0, duration=0.03)
sq_segment.plot()
```

```python
sp = square_wave.make_spectrum()
sp.plot(high=3000)
```

## Low pass filter

```python
sp.low_pass(1000)
sp.plot()
```

```python
square_wave_low_pass = sp.make_wave()
square_wave_low_pass.make_audio()
```

```python
sq_segment = square_wave_low_pass.segment(start=0, duration=0.03)
sq_segment.plot()
```

## High pass filter

```python
sp = square_wave.make_spectrum()
sp.high_pass(1000)
sp.plot()
```

```python
square_wave_high_pass = sp.make_wave()
square_wave_high_pass.make_audio()
```

```python
sq_segment = square_wave_high_pass.segment(start=0, duration=0.03)
sq_segment.plot()
```
